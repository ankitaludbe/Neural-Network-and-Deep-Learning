```python
import numpy as np
import matplotlib.pyplot as plt

def binaryStep(x):
  return np.heaviside(x,1)
```

```python
x=np.linspace(-14,10)
plt.plot(x,binaryStep(x))
plt.axis('tight')
plt.title('Binary Step')
plt.show()
```

```python
def linear(x):
  return x
```

```python
x=np.linspace(-10,10)
plt.plot(x,linear(x))
plt.axis('tight')
plt.title('Linear')
plt.show()
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x)) 
```

```python
x=np.linspace(-14,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Sigmoid')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
x=np.linspace(-14,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('tanh')
plt.show()
```

```python
def RELU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(1)
  return x1
```

```python
x=np.linspace(-14,10)
plt.plot(x,RELU(x))
plt.axis('tight')
plt.title('RELU')
plt.show()
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x=np.linspace(-14,10)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title('Softmax')
plt.show()
```
